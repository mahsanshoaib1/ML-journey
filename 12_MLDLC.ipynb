# <pan style = "color:#004dcf"> Machine Learnning Development Lyfe Cycle(MLDLC) </span>

Its set of guidlines that tells the ML developer from idea to the product. Its complete procedure of make end to end product. There are 9 steps in it:

## <span style = "color:red">1. Frame the problem</span>

**Think about all the questions. Find questions about the project.**
> Define the problem that have to solve. What's the issue that needed to be cure.

> Who is the target. 

> What will be the end product. 

> How many members in the team.

> The cost of project. 

## <span style = "color:red">2. Collection of Data</span>

> Direct CSV file

> From API

> Web Scrapping

> From Database. We take data from the database and make a dataware house

> From clusters. Big data. Spark

## <span style = "color:red">3. Data Preprocessing</span>

**Data we got is always dirty and not able to use.**
> Noisy data

>  Remove Outliers in data, 

> Scalling the data

> Removing Duplicates in data

> Missing values in data

> Shape of data is different.

Make the data compatible for machine learning. 

## <span style = "color:red">4. Exploratory Data Analysis (EDA)</span>

> Exploring the data. 

> Find relationships between intputs and outputs.

> Experiments with data.

> Univariant analysis: analysis of every single variable. std, mean, curves, variance

> Bivariant analysis: Analysis of two variables at a time. In their relation ship

> Multivariant analysis: Analysis of multiple variables, their relationships. 

> Imbalanced data. Make it balanced. For example in image analysis, cat pictures are more than dog pictures. 

## <span style = "color:red">5. Feature Engineering and Selection</span>

> Creating new features from data to make analysis easier. 

> Changing in old columns/features. 

> Relevent Input Features selection. 

## <span style = "color:red">6. Model training, Evaluation, Selection</span>

> Taking multiple algorithms. Training them on data. 

> Use mulitple algorithms. Then evaluate them which is performing better using matrics

> Model Selection. One or more models are selected. Them imputing their parameters. 

> Hyper-Parameter tuning

> Ensemble learning: joining two or more models and making a powerful one. 

## <span style = "color:red">7. Model Development</span>

> Converting model in a software. A website, dashboard, mobile app, dekstop app.    

> Making a binary file of model using some tool eg. pickle

> Then convert this file into an API. API gives jason output.
 
**User input to server to app -> app to API -> API outputs json**

## <span style = "color:red">8. Testing</span>

> Beta testing. A/B testing

> Feedback from user. 

## <span style = "color:red">9. Optimize/span>

> launch model on server.

> Before launching there are a number of steps to be done
> - Backup of model
> - data backup
> - load blamsing. rush on website
> - retraining the model after some time. **Roting.**
