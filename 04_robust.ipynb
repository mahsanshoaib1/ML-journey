## <span style = "color:#b3b258" >Robust Scalling</span>

> **Used when data has outliers in it.**

> X' = X - median / IQR

> IQR = 75th percentile value - 25th percentile value

> Robust to outliers.

```python
#import libraries
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# load dataset
df = sns.load_dataset("diamonds")

# X, y
X = df[["x", "y"]]
y = df["z"]

# train test split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=45)

# apply the model
rbs = RobustScaler()
rbs = rbs.fit(X_train)
rbs_scaled = rbs.transform(X_train)

# converting to dataframe
rbs_scaled = pd.DataFrame(rbs_scaled, columns=X_train.columns)
```

```python
f, axe = plt.subplots(nrows=1,ncols=2, figsize=(12,5))
sns.boxplot(y = X_train["x"], ax = axe[0])
sns.boxplot(y = X_train["y"], ax = axe[1])
```

```python
f,(ax1,ax2) = plt.subplots(ncols=2, figsize=(12,5))
ax1.boxplot(x = rbs_scaled["x"])
ax1.set_title("Before Normalization")
ax2.boxplot(x = rbs_scaled["y"])
ax2.set_title("After normalization")
```
